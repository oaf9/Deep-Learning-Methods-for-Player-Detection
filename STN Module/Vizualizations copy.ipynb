```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
results = pd.read_csv('/Users/omarafifi/Downloads/Jersey Detection/Data/Output from un-attended spt/results.csv')
results.head()
```

```python
fig, ax = plt.subplots(1,2, figsize = (10,5))


fig.suptitle("Metrics for Training over 50 Epochs")
ax[0].plot(np.arange(0,50,1), results['Training Loss'][:50], color = '#9C2F2F', label = 'Training Loss')
ax[0].plot(np.arange(0,50,1), results['Validation Loss'][:50], color = '#343044', label = 'Validation Loss')

ax[1].plot(np.arange(0,50,1), results['Training Accuracy'][:50], color = '#9C2F2F', label = 'Training Accuracy')
ax[1].plot(np.arange(0,50,1), results['Validation Accuracy'][:50], color = '#343044', label = 'Validation Accuracy')
ax[1].set_xlabel('Epoch')
ax[1].set_ylabel('Accuracy %')
ax[0].set_xlabel('Epoch')
ax[0].set_ylabel('Loss')

ax[0].grid()
ax[0].legend()

ax[1].grid()
ax[1].legend()
plt.show()
plt.savefig(fname = '/Users/omarafifi/Downloads/Jersey Detection/Data/Visualization Images/training_metrics.png')
```
